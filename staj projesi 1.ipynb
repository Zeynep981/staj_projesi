```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
# csv türündeki dosyadaki veriler dosya yolu bağlantısı aracılığıyla okutuldu pandas kütüphanesiyle veri çerçevesine veriler aktarıldı.
#head metoduyla ilk 20 satır görüntülendi.
data_set=pd.read_csv('/content/drive/MyDrive/Adsız klasör/dataset.csv')
data_set.head(20)
```

```python
#eksik verilerin tespi edilmesi
eksik_veriler = data_set.isnull().sum()
print("Eksik veriler:")
print(eksik_veriler[eksik_veriler > 0])
```

```python
# Hatalı verilerin kontrol edilmesi ve düzeltilmesi
# Örneğin, Age sütunundaki negatif değerleri kaldırma
data_set = data_set[data_set['Age'] >= 0]

# Veri tiplerinin düzenlenmesi
data_set['Age'] = data_set['Age'].astype(int)
data_set['CreditScore'] = data_set['CreditScore'].astype(int)
```

```python
data_set.head(20)
```

```python
#data_set['Gender'] ile data_Set veri çerçevesindeki cinsiyet bilgilerini içeren sütun seçildi,value_counts() metodu ile
#seçilen sütundaki değerlerin her birinin kaç kez tekrar ettiği sayıldı,bu bilgiler top_gender değişkenine atanarak print
#ile ekrana yazıldı.
top_gender=data_set['Gender'].value_counts()
print(top_gender)
```

```python
#bir önceki top_gender'dan farklı olarak index[0] eklendi bu sayede value.counts metodu ile
#oluşturulan dizindeki 0. indeksli yani en büyük(en çok tekrar eden) olan elemana erişim sağlanıldı.ve bu eleman ekrana yazıldı.
top_age=data_set['Age'].value_counts().index[0]
print(" listede en çok bulunan yaş:",top_age)
```

```python
#oluşturulan dizindeki 0. indeksli yani en büyük(en çok tekrar eden) olan elemana erişim sağlanıldı.ve bu eleman ekrana yazıldı.
top_geography=data_set['Geography'].value_counts().index[0]
print("en çok yer alan ülke:",top_geography)
```

```python
#data_set['Geography'] == 'Spain'' ile nobel veri çerçevesinin geography sütunundaki değerler kontrol edilir.
#her bir kişinin geography verisinin spain olup olmadığını kontrol eder . sonuçlar boolean serisi şeklindedir.
#Bu seri, "italyanlar" adlı bir değişkene atanır ve bu değişken, her bir kişinin italyan
#olup olmadığını belirten bir seriyi içerir.
italyanlar=data_set['italyanlar'] = data_set['Geography'] == 'Spain'
print(italyanlar)
```

```python
#altta yer alan ifade kredi skor değerlerini kullanarak onluk skor  adında yeni bir sütun oluşturur
#np.floor(data_set['CreditScore'] / 10) * 10 ile her skor onlar basamağına yuvarlanır bu işlem sonucunda her skorun
#değer aralığı 10  olarak belirienir.
data_set['onluk_skor'] = (np.floor(data_set['CreditScore'] / 10) * 10)
italyanların_skor_ortu = data_set.groupby('onluk_skor', as_index=False)['italyanlar'].mean()
#yukarıda yer alan ifade nobel veri çerçevesini on_yıllık_zaman sütununa göre gruplar as_index=False  ile
#gruplama sırasında index oluşturulmaz ['amerikalı_kazananlar'].mean() ile amerşkalı ödül kazananların yıllık
#ortalaması hesaplanır ve amerikalı_kazananların_ortu değişkenine kaydedilir.
print(italyanların_skor_ortu)
```

```python
#altta yer alan kodlarda max() metodu aracılığıyla italyanların en yüksek ortalaması bulunur.
#['onluk_skor'].values[0] kısmı ile skor ortalaması en yüksek olan satırların onluk skor sütunundaki
#değerleri alınır .values[0] kısmı ile elde edilen değer dizi şeklinde değil de tek bir değer olarak değer alır
#elde edilen değer max_skor_spain değişkenine atanır
max_skor_spain = italyanların_skor_ortu [italyanların_skor_ortu['italyanlar'] ==
                                  italyanların_skor_ortu['italyanlar'].max()]['onluk_skor'].values[0]
print(max_skor_spain)
```

```python
#altta yer alan ifade kredi skor değerlerini kullanarak onluk skor  adında yeni bir sütun oluşturur
#np.floor(data_set['CreditScore'] / 10) * 10 ile her skor onlar basamağına yuvarlanır bu işlem sonucunda her skorun
#değer aralığı 10  olarak belirienir.
data_set['onluk_skor'] = (np.floor(data_set['CreditScore'] / 10) * 10)
#altta yer alan kodlarda dataset veri çerçevesine kazanan kadınlar adında yeni bir sütun eklenir.
#bu sütun dosyada yer alan her kişinin kadın olup olmadığını test eder. elde edilen verilen bir boolean serisidir.
#bu serideki veriler skor_kadınlar değişkenine atanır.
skor_kadınlar = data_set['skor_kadınlar'] = data_set['Gender'] == 'Female'
#alttaki kodlarda dataset veri çerçevesi onluk skor ve yaş sütunlarına göre gruplandırılır.
#ardından elde edilen her bir grup için skor kadınlar sütununun ortalaması alınır.ve bu bilgiler
#skor_kadınlar_ortu adlı değişkene atanır.
skor_kadınlar_ortu = data_set.groupby(['onluk_skor', 'Age'], as_index=False)['skor_kadınlar'].mean()
print(skor_kadınlar.head(10))

```

```python
print(f"\n kadınların ortalama skoru:",skor_kadınlar_ortu)
```

```python
#.max() metodu kullanılarak skor kadınlar ortalamasının en yüksek olduğu satırar seçilir.
#[['onluk_skor', 'Age']] kısmı ile elde edilen maks ortalamaların sadece onluk skor ve yaş ile ilişkili
#sütunları seçilmiş olur. elde edilen değerler max_kadın_skor adlı değişkene atanır.
max_kadın_skor = skor_kadınlar_ortu[skor_kadınlar_ortu['skor_kadınlar'] ==
                                        skor_kadınlar_ortu['skor_kadınlar'].max()][['onluk_skor', 'Age']]
print(max_kadın_skor)
```

```python
#max_kadın_skor veri çerçevesindeki sırasıyla onluk skor ve yaş sütunları ile ilgili
#değerler alınır values[0] kısmıyla değerler dizi olarak değil de değer olarak saklanır.
#onluk skor anahtar,yaş ise değer olarak atanır ve sözlük oluşturulur.
max_kadın_dict = {max_kadın_skor['onluk_skor'].values[0]: max_kadın_skor['Age'].values[0]}
print(max_kadın_dict)
```

```python
#data set veri çerçevesinden gender sütunundaki değerlerin female olma durumunu kotrol edip,
#elde edilen boolean serisi türündeki değerleri skor kadınlar değişkenine atar.
skor_kadınlar = data_set['skor_kadınlar'] = data_set['Gender'] == 'Female'
#ds_kadınlar adında yeni bir veri çerçevesi oluşturulur.skor_kadınlar sütununu
#içeren satırları içerir.
ds_kadınlar = data_set[data_set['skor_kadınlar']]
#ilk_deger adında yeni bir veri çerçevesi oluşturulur..min() metodu sayesinde
#ds_kadınlar veri çerçevesindeki en küçük değere sahip çalışma süresi satırlarını içerir.
ilk_deger = ds_kadınlar[ds_kadınlar['Tenure'] == ds_kadınlar['Tenure'].min()]
#ilk_deger veri çerçevesindeki Surname sütunundaki değerlere erişir ve .values[0] ile ilk
#değeri seçer. Yani, en az süre çalışan kadının soy adını (Surname) alır.
first_woman_sname = ilk_deger['Surname'].values[0]
#ilk_deger veri çerçevesindeki yaş sütunundaki değerlere erişir ve .values[0] ile ilk
#değeri seçer. Yani, en az süre çalışan kadının yaşını alır.
first_woman_age = ilk_deger['Age'].values[0]
```

```python
#data set veri çerçevesindeki Surname sütunundaki değerlerin value.counts() metodunu kullanarak sayısını hesaplar ve
#her bir değeri, o değerin kaç kez tekrar ettiğini içeren bir Seri oluşturur.
deger = data_set['Surname'].value_counts()
#deger Serisi içindeki değerleri kontrol edilir ve en az 2 kez tekrar edenleri seçilir.
#.index ile de bu tekrar edenlerin indekslerini alınmış olur.
tekrar_edenler = deger[deger >= 2].index
# tekrar_edenler serisi içindeki indeksler bir listeye dönüştürülüp repeat_list değişkenine atanır.
repeat_list = list(tekrar_edenler)
print(deger)
print(tekrar_edenler)
print("tekrar eden soyadlar  :", repeat_list)
```

```python
plt.figure(figsize=(10, 6))
#Bu satır, grafik penceresinin boyutunu 10x6 inç olarak ayarlar. plt.figure fonksiyonu ile yeni bir figür
#oluşturulur ve figsize parametresi ile boyutlandırılır.

sns.countplot(x='Gender', data=data_set, palette=['red'])
#bu satır, seaborn kütüphanesini kullanarak bir çubuk grafiği oluşturur. sns.countplot fonksiyonu,
#'Gender' sütunundaki kategorilere göre (örneğin, 'Erkek' ve 'Kadın') veri setindeki müşteri sayılarını sayar
#ve bir çubuk grafik oluşturur. x='Gender' parametresi, x ekseninde 'Gender' değerlerin kullanılacağını belirtir
#ve data=data_set parametresi ise kullanılan veri setini belirtir.
#palette=['red'] parametresi, çubukların renginin kırmızı olmasını sağlar.

plt.title('Cinsiyete Göre Müşteri Sayısı')
plt.xlabel('Cinsiyet')
plt.ylabel('Müşteri Sayısı')
#Bu üç satır, grafiğin başlığını ve eksen etiketlerini ayarlar. plt.title fonksiyonu grafiğe bir başlık ekler,
#plt.xlabel fonksiyonu x ekseninin etiketini belirler ve plt.ylabel fonksiyonu y ekseninin etiketini belirler.

plt.show()
#Bu satır, oluşturulan grafiği ekranda görüntüler. plt.show fonksiyonu, matplotlib figürünü ekranda görüntülemek için kullanılır.
```

```python
plt.figure(figsize=(10, 6))
# Bu satır, grafik penceresinin boyutunu 10x6 inç olarak ayarlar. plt.figure fonksiyonu ile yeni bir figür
# oluşturulur ve figsize parametresi ile boyutlandırılır.

sns.scatterplot(x='Age', y='CreditScore', hue='Exited', data=data_set, palette='coolwarm', alpha=0.5)
# Bu satır, seaborn kütüphanesini kullanarak bir dağılım grafiği oluşturur. sns.scatterplot fonksiyonu,
# 'Age' ve 'CreditScore' sütunlarındaki verilere göre bir dağılım grafiği çizer. x='Age': X ekseninde 'Age' sütunundaki değerler kullanılır.
# y='CreditScore': Y ekseninde 'CreditScore' sütunundaki değerler kullanılır.
# hue='Exited': Verileri 'Exited' sütunundaki değerlere göre renklendirir. Yani, müşterinin bankadan ayrılıp
# ayrılmadığını (0 veya 1) renklerle gösterir. data=data_set: Kullanılan veri setini belirtir.
# palette='coolwarm': Renk paletini 'coolwarm' olarak ayarlar, yani ayrılma durumu olan ve olmayan müşterileri  farklı renklerle gösterir.
# alpha=0.5: Noktaların saydamlık seviyesini %50 olarak ayarlar, yani noktalar biraz saydam olur.

plt.title('Yaş ve Kredi Skoru Arasındaki İlişki')
plt.xlabel('Yaş')
plt.ylabel('Kredi Skoru')
# Bu üç satır, grafiğin başlığını ve eksen etiketlerini ayarlar. plt.title fonksiyonu grafiğe bir başlık ekler,
# plt.xlabel fonksiyonu x ekseninin etiketini belirler ve plt.ylabel fonksiyonu y ekseninin etiketini belirler.

plt.show()
# Bu satır, oluşturulan grafiği ekranda görüntüler. plt.show fonksiyonu, matplotlib figürünü ekranda görüntülemek için kullanılır.
```
